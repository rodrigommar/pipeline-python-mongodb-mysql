## Desafio: Transformar alguns dados com objetivo de melhorar a **análise** e gerar **insights**

```python
import pandas as pd
```

```python
# ler arquivo "banco_clientes.csv" e gerar um dataframe 

df = pd.read_csv('../data/banco_clientes.csv')
```

```python
df.info()
```

```python
df
```

# Por o nome em LETRA MAIUSCULA

```python
df['nome'] = df['nome'].str.upper()
df.head()
```

# Padronizar a coluna telefone

* Remover os parenteses do dado telefone.
* Formato: dd ddddd-dddd

```python
df.telefone.str.replace('(', '', regex=False)
```

```python
# remove o '(' 
# .str.replace('(', '', regex=False)

# remove o ')
# .str.replace(')', '', regex=False)

df['telefone'].str.replace('(', '', regex=False).str.replace(')', '', regex=False)
```

```python
df['telefone'] = df['telefone'].str.replace('(', '', regex=False).str.replace(')', '', regex=False)
df
```

# Adiciona uma coluna CEP no dataframe

```python
df.endereco.str.extract('(\d{5}-\d{3})')
```

```python
df['cep'] = df.endereco.str.extract('(\d{5}-\d{3})')
df
```

```python
df.endereco.str.replace('\d{5}-\d{3}', '', regex=True)
```

```python
df['endereco'] = df.endereco.str.replace(r'\d{5}-\d{3}', '', regex=True).str.strip()
```

```python
df
```

# Criar uma coluna para Rua

```python
df.endereco.str.split('\s(?=\d+$)', expand=True)
```

```python
df.endereco.str.replace(',', '').str.split('\s(?=\d+$)', expand=True)
```

```python
df[['rua', 'numero']] = df.endereco.str.replace(',', '').str.split('\s(?=\d+$)', expand=True)
```

```python
df
```

```python
# axis=1 , as colunas serão removidas
# axis=2 , as linhas serão removidas 

df = df.drop(['endereco'], axis=1)
df
```

# Salvando uma nova versão do dataframe tratado em um arquivo csv

```python
df.to_csv('../data/banco_clientes_editado.csv')
```
